```python
import pandas as pd
import numpy as np
import json
import requests
import pycountry
import openweathermapy.core as owm
from citipy import citipy as cp
from config import api_key
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
lat=np.random.uniform(low=-66, high=66, size=10)
lon=np.random.uniform(low=-180, high=180, size=10)
zip_cord=zip(lat,lon)
city=[]
lat=[]
lon=[]
name_country=[]
for x,y in zip_cord:
    geo = cp.nearest_city(x,y)
    city.append(geo.city_name.title())
    #Capturing Lat to Lon to identify if weather map have details about the city or not
    lat.append(x)
    lon.append(y)
    name_country.append(pycountry.countries.get(alpha_2=geo.country_code.upper()).name)


cities_df=pd.DataFrame({"City":city,"Country":name_country,"Latitude":lat,"Longitude":lon })
cities_df["Country Code"]=""
#cities_df["Latitude"]=""
#cities_df["Longitude"]=""
cities_df["Max Temp"]=""
cities_df["Humidity"]=""
cities_df["Cloudiness"]=""
cities_df["Wind Speed (mph)"]=""
cities_df["Date"]=""
cities_df["Time"]=""
cities_df
```

```python
cities_df
```

```python
settings = {"units": "imperial", "appid": api_key}
for index,row in cities_df.iterrows():
    try:
        weather_city = owm.get_current(row["City"], **settings)
        print(f"Current weather object: {weather_city}.")
        if(weather_city):
        #row["Country"]=weather_city["sys"]["country"]
            cities_df.set_value(index, "Latitude",weather_city["coord"]["lat"]  )
            cities_df.set_value(index, "Longitude",weather_city["coord"]["lon"] )
            cities_df.set_value(index, "Wind Speed (mph)",weather_city['wind']['speed'] )
            cities_df.set_value(index, "Max Temp", weather_city["main"]["temp_max"])
            cities_df.set_value(index, "Humidity",weather_city["main"]["humidity"] )
            cities_df.set_value(index, "Cloudiness",weather_city["clouds"]["all"] )
            cities_df.set_value(index, "Date", datetime.fromtimestamp(weather_city["dt"]).strftime("%m/%d/%Y"))
            cities_df.set_value(index, "Time", datetime.fromtimestamp(weather_city["dt"]).strftime("%I:%M:%S"))
            cities_df.set_value(index, "Country Code", weather_city["sys"]["country"])
            cities_df.set_value(index, "Country",pycountry.countries.get(alpha_2=weather_city["sys"]["country"]).name )
    except:
        cities_df.drop(index,inplace=True)
        continue


cities_df   
```

```python
cities_df['Latitude']=pd.to_numeric(cities_df['Latitude'])
cities_df['Max Temp']=pd.to_numeric(cities_df['Max Temp'])
cities_df['Humidity']=pd.to_numeric(cities_df['Humidity'])
cities_df['Cloudiness']=pd.to_numeric(cities_df['Cloudiness'])
cities_df['Wind Speed (mph)']=pd.to_numeric(cities_df['Wind Speed (mph)'])

cities_df.head()
```

```python
#plt.scatter(cities_df["Latitude"], "Temperature(F)"], marker="o", facecolors="red", edgecolors="black", alpha=0.75)
#cities_df.plot(kind="scatter",x="Latitude",y="temp_max",marker="o")

with plt.style.context('seaborn-dark'):
    plt.title(f"City Latitude vs Max Temperature ({cities_df['Date'][0]} {cities_df['Time'][0]})")
    plt.xlim(-90,90)
    plt.ylim(min(cities_df['Max Temp'])-10,max(cities_df['Max Temp'])+10)
    plt.xlabel("Latitude")
    plt.ylabel("Max Temperature (F)")
    plt.grid(color='white', linestyle='-', linewidth=1)
    plt.scatter(cities_df["Latitude"], cities_df["Max Temp"],edgecolors="black", color="blue", alpha=.8)
```
